```python
import pandas as pd
import os
import csv
import sys
import numpy as np

data_file = os.path.join(os.path.expanduser('~'), 'Downloads', 'NBA.csv')

data = pd.read_csv(data_file, sep='\,', header=1,engine='python')

data
```

```python
data['Notes"'].unique()
```

```python
columnsDel = data.columns[-5:]
columnsDel
columnsDel2 = list(columnsDel)
columnsDel2
```

```python
data.drop(columnsDel2, axis=1, inplace= True)
```

```python
data
```

```python
data.columns = ['Date', 'Time','Visitor Team', 'VisitorPts','Home Team','HomePts','OT?','Notes','Attend' ]
```

```python
data
```

```python
data['Date']= data['Date'].map(lambda x: x[4:])
```

```python
data['Date'] = pd.to_datetime(data['Date'], infer_datetime_format=True)
```

```python
data.shape
```

```python
data['HomeWin'] = data['HomePts'] > data['VisitorPts']

y_true = data['HomeWin'].values
```

```python
y_true
```

```python
data.head()
```

```python
data['HomeLastWin'] = False
data['VisitorLastWin'] = False
```

```python
# 创建两个特征，分别是两支球队上场比赛的胜负情况
# 创建默认字典，储存球队上次比赛的结果

from collections import defaultdict
won_last = defaultdict(int)

for index, row in data.sort_values(by=('Date')).iterrows(): #该方法并不高效
    # 提取队伍名称
    home_team = row['Home Team']
    visitor_team = row['Visitor Team']
    
    # 将LastWin的值，赋给lastWin列
    row['HomeLastWin'] = won_last[home_team]
    row['VisitorLastWin'] = won_last[visitor_team]
    data.loc[index] = row
    
    # 设定当前赢
    won_last[home_team] = row['HomeWin']
    won_last[visitor_team] = not row['HomeWin']
    
    #标记的是，主场与客场球队上一轮的胜负结果。
```

### 使用决策树进行分析

```python
# 使用决策树进行分析

from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state = 1)

# 抽取指定列数据

X_previouswins = data[['HomeLastWin', 'VisitorLastWin']].values
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X_previouswins, y_true, scoring='accuracy',cv=10)
scores
```

### 使用以往战绩预测胜率

```python
# 使用以往战绩预测胜率

standings_file = os.path.join(os.path.expanduser('~'), 'Downloads', 'NBA Standings.csv')
standingsData = pd.read_csv(standings_file, skiprows=[0])
```

```python
standingsData.head()
```

```python
data['HomeTeamRanksHigher'] = 0

for index, row in data.iterrows():
    home_team = row['Home Team']
    visitor_team = row['Visitor Team']
    
    if home_team == "New Orleans Pelicans":
        home_team = "New Orleans Hornets"
    elif visitor_team == "New Orleans Pelicans":
        visitor_team = "New Orleans Hornets"
        
    home_rank = standingsData[standingsData['Team']== home_team]['Rk'].values
    visitor_rank = standingsData[standingsData['Team']== visitor_team]['Rk'].values
    
    row['HomeTeamRanksHigher'] = int(home_rank > visitor_rank)
    data.iloc[index] = row
```

```python
data.iloc[20:25]
```

```python
X_homehigher = data[['HomeLastWin', 'VisitorLastWin', 'HomeTeamRanksHigher']].values
```

```python
clf = DecisionTreeClassifier(random_state = 1)
scores = cross_val_score(clf, X_homehigher, y_true, scoring='accuracy', cv=10)
```

```python
print('The accuracy is {0:.2f}%'.format(np.mean(scores)*100))
```

```python
from sklearn.model_selection import GridSearchCV

parameter = {'max_depth':list(np.arange(1,21))}

clf = DecisionTreeClassifier(random_state = 1)
grid = GridSearchCV(clf, parameter, cv=10)

grid.fit(X_homehigher, y_true)
```

```python
print("The accuracy is {0:.2f}%".format(grid.best_score_ *100))
```

### 比较对战两支球队，上次交手的胜负关系

```python
# 比较对战两支球队，上次交手的胜负关系

last_match_winner = defaultdict(int)
data['HomeTeamWonLast'] = 0

for index, row in data.sort_values(by=('Date')).iterrows():
    home_team = row['Home Team']
    visitor_team = row['Visitor Team']
    
    #不考虑主客场作战。按照英文字母表对球队名字进行排序，确保两个球队无论主客场，都使用相同的键
    teams = tuple(sorted([home_team, visitor_team]))
    
    row['HomeTeamWonLast'] = 1 if last_match_winner[teams] == row['Home Team'] else 0
    
    winner = row['Home Team'] if row['HomeWin'] else row['Visitor Team']
    data.iloc[index] = row
    
    last_match_winner[teams] = winner
    
```

```python
X_lastwinner = data[['HomeTeamWonLast','HomeTeamRanksHigher']].values

clf = DecisionTreeClassifier(random_state = 1)
grid = GridSearchCV(clf, parameter, cv=10)

grid.fit(X_lastwinner, y_true)
print('The accuracy is {0:.2f}%'.format(grid.best_score_*100))
```

```python
score = cross_val_score(clf, X_lastwinner, y_true, scoring='accuracy',cv=10)
print('The accuracy is {0:.2f}%'.format(np.mean(score)*100))
```

### 将球队信息加入决策树

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# 将主场球队名称转化为整型

encoder = LabelEncoder()
encoder.fit(data['Home Team'].values)

#抽取所有主客场的球队名称，将其组合

home_teams = encoder.transform(data['Home Team'].values)
visitor_teams = encoder.transform(data['Visitor Team'].values)

X_teams = np.vstack([home_teams, visitor_teams]).T
```

```python
onehot = OneHotEncoder()
X_teams = onehot.fit_transform(X_teams).todense()
```

```python
clf = DecisionTreeClassifier(random_state = 1)
score = cross_val_score(clf, X_teams, y_true, scoring='accuracy', cv=10)
```

```python
print("The accuracy is {0:.2f}%".format(np.mean(score)*100))
```

### 使用随机森林

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1)
scores = cross_val_score(clf, X_teams, y_true, scoring='accuracy',cv=10)
print('The accuracy is {0:.2f}%'.format(np.mean(scores)*100))
```

```python
#使用不同的特征子集进行学习

X_all = np.hstack([X_homehigher, X_teams])
clf = RandomForestClassifier(random_state=1)
scores = cross_val_score(clf, X_all, y_true, scoring='accuracy', cv=10)
print('The accuracy is {0:.2f}%'.format(np.mean(scores)*100))
```

```python
# 使用GridSearchCV 搜索最佳参数

parameter_space = {
    'max_features':[2,10, 'auto'],
    'n_estimators':[100,],
    'criterion':['gini', 'entropy'],
    'min_samples_leaf':[2,4,6]
}

clf = RandomForestClassifier(random_state=1)
grid = GridSearchCV(clf, parameter_space)
grid.fit(X_all, y_true)

print("The accuracy is {0:.2f}%".format(grid.best_score_*100))
```
